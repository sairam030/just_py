## Funtions in Python

A funtion is a block of code that performs a specific task. Funtions help in organizing code, reusing code,and improving readability.

```python
## syntax
def funtion_name(parameters):
  """Docstring"""
  # Funtion body
  return expression
```

```python
## why functions?

num=24
if num%2==0:
  print("the number is even")
else:
  print("the number is odd")
```

```python
def even_odd(num):
  """ This fun finds even or odd """
  if num%2==0:
    print("the number is even")
  else:
    print("the number is odd")
```

```python
## call function

even_odd(56)
```

```python
 ## function with multiple parameter

def add(a,b):
  c=a+b
  return c

result=add(2,4)
print(result)
```

```python
## Default parameters

def greet(name):
  print(f"Hello {name} Welcome")

greet("Ram")
```

```python
def greet(name="Guest"):
  print(f"Hello {name} Welcome")

greet()
```

```python
### varibale lenght arguments
## Positional and keywords arguments

def print_numbers(*args): ## *anything
  for numbers in args:
    print(numbers)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Ram")
```

```python
## keywords arguments

def print_details(**kwargs):
  for key,val in kwargs.items():
    print(f"{key}:{val}")
```

```python
print_details(name="ram",age=23,grade="a")
```

```python
def print_details(*args,**kwargs):
  for val in args:
    print(f"Positional:{val}")

  for key,val in kwargs.items():
    print(f"{key}:{val}")
```

```python
print_details(1,2,3,4,5,name="ram",age=23,grade="a")
```
