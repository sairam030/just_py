### The map() funtion in Python

The map() function applies a given funtion to all items in an input list (or any other iterable) and returns a map object (an iterator). This is particularly useful for transforming data in a list comprehensively.

```python
def square(x):
  return x**2

square(4)
```

```python
numbers=[1,2,3,4,5,6]

list(map(square,numbers))
```

```python
## lambda funtion with map

numbers=[1,2,3,4,5,6]

list(map(lambda x:x**2 ,numbers))
```

```python
### map mutliple iterables

num1=[1,2,3]
num2=[4,5,6]

added=list(map(lambda x,y:x+y,num1,num2))
added
```

```python
## map() to convert a list of str to intergers

str_n=['1','2','3','4','5']
int_n=list(map(int,str_n))
print(int_n)
print(str_n)
```

```python
words =['apple','banana','cherry']
upper_word=list(map(str.upper,words))
upper_word
```

```python
def get_name(person):
  return person['name']

people=[{'name':"ram","age":23},{'name':"sai","age":32}]

list(map(get_name,people))
```
