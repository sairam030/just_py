### Lambda Functions in Python

Lambda functions are small anonymouus defined using the Lambda keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions

```python
## Syntaxx

lambda arguments: expression
```

```python
def addition(a,b):
  return a+b
```

```python
addition(2,4)
```

```python
addition_l=lambda a,b:a+b
type(addition_l)

print(addition_l(5,6))
```

```python
def even(num):
  if num%2==0:
    return True

even(34)
```

```python
even1 = lambda num:num%2==0
even1(6)
```

```python
def addition(x,y,z):
  return x+y+z

addition(1,3,5)
```

```python
addition1 = lambda a,b,c:a+b+c
addition1(2,3,4)
```

```python
## map() - applies a fun to all items in a list

numbers=[1,2,3,4,5,6]

def square(number):
  return number**2

square(4)
```

```python
list(map(lambda x:x**2,numbers))
```
