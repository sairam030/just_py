### Dictionaries

Dict are unorderd collections of items. They store data in key-value pairs.Keys must be unique and immutable(eq strings numbers or tuples),while values can be of any type.

```python
## Creating Dict

empty_dict={}
type(empty_dict)
```

```python
empty_dict=dict()
empty_dict
```

```python
students={"name":"ram","age":32,"grade":24}
students
```

```python
## Error single key is used
students={"name":"ram","age":32,"grade":24,"name":32}
students
```

```python
## accesing Dict elements
students={"name":"ram","age":32,"grade":"A"}
students
```

```python
students['grade']
students['age']
```

```python
## using the get()

print(students.get("grade"))
print(students.get("last_name"))

print(students.get("last_name","NOT AVBL"))
```

```python
## modifing dict elemts
## dict are mutable, so you can add, update or delete elements

print(students)

students["age"]=32 ## update
print(students)

students["add"]="India" ## add
print(students)
```

```python
del students["grade"] ## del key val pair

print(students)
```

```python
## Dict methods

keys=students.keys() ## all keys
print(keys)

values=students.values() ## all values
print(values)

items= students.items() ## all
print(items)
```

```python
## shallow copy // deep copy
student_copy=students
print(students)
print(student_copy)
```

```python
students["name"]="SAI"

print(students)
print(student_copy)
```

```python
student1=students.copy()

students["name"]="Ram"

print(students)
print(student1)
```

```python
## iterating over dict
## for keys
for keys in students.keys():
    print(keys)
```

```python
## for values
for val in students.values():
    print(val)
```

```python
## for key-value

for key,val in students.items():
    print(f"{key}:{val}")
```

```python
## nested dict
students={
    "student1":{"name":"Ram","age":23},
    "student2":{"name":"Sai","age":21}
}
print(students)
```

```python
## access nested dict
print(students["student2"]["name"])
```

```python
 ## Iterating over nested dict

for students_id,student_info in students.items():
  print(f"{students_id},{student_info}")
  for key, val in student_info.items():
    print(f"{key},{val}")
```

```python
## Dictinary comprehention

squares={x:x**2  for x in range(5)}  ## {key:val condition}
print(squares)
```

```python
## conditional dict comprehension

even = {x:x**2 for x in range(10) if x%2==0}
even
```

```python
#### Examplesss

## use dict to count a freq of elements in list

num=[1,2,2,3,3,3,4,4,4,4]
count=0
dit={}
for x in num:
  if x in dit:
    count=count+1
  else:
    count=1
  dit[x]=count
print(dit)
```

```python
## merge 2 dict

dict1={"a":1,"b":2}
dict2={"b":3,"c":3}

merge={**dict1,**dict2}
merge
```
